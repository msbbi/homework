```python
import numpy as np
import matplotlib.pyplot as plt

# 定義方程式的係數
a = -2
b = 4
c = 5

# 生成 x 值的範圍
x = np.linspace(-10, 10, 100)

# 計算對應的 y 值
y = a * x**2 + b * x + c

# 繪製圖形
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = {}x^2 + {}x + {}'.format(a, b, c))
plt.grid(True)
plt.show()
```

```python
# 定義方程式的係數
a = -2
b = 4
c = 5

# 計算最高點的 x 坐標
x_vertex = -b / (2 * a)

# 計算最高點的 y 坐標
y_vertex = a * x_vertex**2 + b * x_vertex + c

# 輸出最高點的坐標
print("最高點的坐標：({:.2f}, {:.2f})".format(x_vertex, y_vertex))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義方程式的係數
a = 3
b = 2
c = -6

# 生成 x 值的範圍
x = np.linspace(-10, 10, 100)

# 計算對應的 y 值
y = a * x**2 + b * x + c

# 繪製圖形
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = {}x^2 + {}x + {}'.format(a, b, c))
plt.grid(True)
plt.show()
```
